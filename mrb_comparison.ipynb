```python
from qiskit_device_benchmarking.bench_code.mrb import MirrorRB, MirrorRBnew
from qiskit.providers.fake_provider import GenericBackendV2
```

```python
seed = 1984

coupling_map = [(0,1), (0,2), (1,3), (2,3)]
backend = GenericBackendV2(4, coupling_map=coupling_map)
```

```python


exp = MirrorRB(
    range(backend.num_qubits),
    lengths=[2,4],
    backend=backend,
    two_qubit_gate_density=0.25,
    num_samples=20,
    seed=seed
    )

print(exp._sample_sequences())

exp.circuits()[0].draw(fold=-1)
```

```python
exp = MirrorRBnew(
    range(backend.num_qubits),
    lengths=[2,4],
    backend=backend,
    two_qubit_gate_density=0.25,
    num_samples=20,
    seed=seed,
    )

print(exp._sample_sequences())

exp.circuits()[0].draw(fold=-1)
```
